Code block was applied from Code Reuse Session 1 - Question 19 and Tute 1 - 31st page

```python
#import library to find outliers in dataset
import numpy as np

#create a funtion to check outliers in each variable
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers
#check outliers for numerical variables
numerical_cols = ['Age', 'Grade', 'Tumor_Size', 'Regional_Node_Examined', 'Reginol_Node_Positive', 'Survival_Months']

for col in numerical_cols:
    outliers = detect_outliers_iqr(data, col)
    print(f"Outliers in {col}: {len(outliers)}")

    if len(outliers) > 0:
      print(outliers[[col]])
    print("\n")
```
