Code block was applied from Tute 4 - 11-13 pages

```python
#tune the hyperparameters and displaying best combination of hyperparametsfor KNN
import numpy as np

#create a new classifier for KNN
knn = KNeighborsClassifier()

#creating a dictionary of values that should be tested
param_grid = {'n_neighbors': np.arange(1, 25), 'metric':['euclidean', 'manhattan']}

#use gridsearch to test all values
knn_cv = GridSearchCV(knn, param_grid, cv=5, scoring='roc_auc')

#fit the data into the model
knn_cv.fit(X_train, y_train)
```

```python
#check top performing n_neighbors value
print("Best Parameters: ", knn_cv.best_params_)
best_knn = knn_cv.best_estimator_
```

```python
#perform predictions on the test dataset with newly formed hyper tuned model
y_pred = best_knn.predict(X_test)

#create the confustion matrix
cm = confusion_matrix(y_test, y_pred, labels=knn_cv.classes_)
disp_knn = ConfusionMatrixDisplay(cm, display_labels=knn_cv.classes_)
disp_knn.plot()
```

```python
# tuning the hyperparameters (max_iter) for the next best model - LR
log_reg = LogisticRegression(class_weight='balanced', solver='liblinear', verbose=1)

param_grid = {
    'C': [0.0001, 0.001, 0.01, 0.1, 1, 10, 100],
    'max_iter': [100, 500, 1000, 2000, 4000]
}

# set up gridsearch cv
log_reg_gscv = GridSearchCV(estimator=log_reg, param_grid=param_grid, cv=5, n_jobs=-1, scoring='f1')
log_reg_gscv.fit(X_train, y_train)

# Best model and parameters
print("Best parameters for Logistic Regression:", log_reg_gscv.best_params_)

# Best cross-validated score
print("Best cross-validated F1 score:", log_reg_gscv.best_score_)
best_lr = log_reg_gscv.best_estimator_
best_lr = log_reg_gscv.best_estimator_

# fit the model
best_lr.fit(X_train, y_train)

# make predictions
y_pred = best_lr.predict(X_test)
```
